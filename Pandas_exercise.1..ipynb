```python
# 영화 영화

import numpy as np
import pandas as pd

df_movies = pd.read_csv('./data/movies.csv')

df_rate = pd.read_csv('./data/ratings.csv')

display(df_movies)
display(df_rate)

 
```

```python
# 1. 사용자가 평가한 모든 영화의 전체 평균 평점

df_rate = pd.merge(df_movies, df_rate, on='movieId', how='inner')

display(df_rate)

#rate_movie = df_rate['rating'].groupby(df_rate['movieId'])

print(df_rate['rating'].mean())

# 3.501556983616962
```

```python
# 2. 각 사용자별 평균 평점

M_rate = (df_rate['rating'].groupby(df_rate['userId']).mean())

print(M_rate)
```

```python
# 3. 각 영화별 평균 평점

df_rate = pd.merge(df_movies, df_rate, on='movieId', how='right')

M_rate_movie = (df_rate['rating'].groupby([df_rate['movieId'],df_rate['title']]).mean())

print(M_rate_movie)


# print(type(M_rate_movie))
# <class 'pandas.core.series.Series'>
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목(동률이 있을 경우 모두 출력)

#print(M_rate_movie)

sorted_list = M_rate_movie.sort_values(ascending=False)

# print(sorted_list)

print(sorted_list[sorted_list == 5.0])
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목

contains_Comedy = df_rate['genres'].str.contains('Comedy')

df_Comedy = df_rate[contains_Comedy]

# display(df_Comedy)

# 평점 기준으로 정렬한 뒤 타이틀, 평점 열만 팬시 인덱싱
sorted_df_Comedy = (df_Comedy.sort_values(by=['rating']).loc[:,['title','rating']])

display(sorted_df_Comedy) # 가장 낮은 평점이 0.5점

display(sorted_df_Comedy.loc[:,'title'][sorted_df_Comedy.loc[:,'rating'] == 0.5])
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?

# movies.csv파일에는 평점을 부여한 모든 영화에 대한 정보를 가지고 있다는 가정

# 장르에 로맨스를 포함한 boolean mask 추출
contains_Romance = df_rate['genres'].str.contains('Romance')

# 장르에 로맨스를 포함한 DataFrame
df_Romance = df_rate[contains_Romance].loc[:,['movieId','title','genres','userId','rating','timestamp']]

display(df_Romance)

# timestamp를 문자열로 변환하기 위해 시간 관련 함수 import
import time

# tm = time.localtime(964981247)
# string_year = time.strftime('%Y', tm)
# print(string_year)

tempArr = list()

for S in df_Romance.loc[:,'timestamp'].values:
    
    tm = time.localtime(S)
    string_year = time.strftime('%Y', tm)
    tempArr.append(string_year)

df_Romance.loc[:,'timestamp'] = tempArr

# timestamp를 년도로 바꾼 df
df_appen_yr_Rom = df_Romance


# 여기서 변형시킨 년도는 '문자열'
# boolean indexing + mean() 함수를 사용했다

print((df_appen_yr_Rom.loc[:,'rating'][df_appen_yr_Rom.loc[:,'timestamp'] == '2015']).mean())

# 3.396375098502758
    
```
