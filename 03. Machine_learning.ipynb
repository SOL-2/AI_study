```python
# 수치미분(Numenrcal derivative)


# 어떤 함수의 정의역 속 각 점에서 
# 독립변수의 값의 변화량과 함수값의 변화량 비율의 극한 혹은 극한의 집합
# 치역으로 가지는 새로운 함수

# 이런 새로운 함수를 구할 수 있는데 이를 미분이라고 하고 derivative라고 한다.
# 악간 다른 의미로 이런 미분함수(도함수)를 구하는 작업을 미분법
# differentiation이라고 한다


# 미분은 함수에 있어서 특정 순간의 변화량을 의미
# x값에서의 작은 변화가 함수 f(x)를 얼마나 변화시키는가요?

# 논리적 전개로 미분을 수행 
# => 해석미분(Anaytical differentiation)

# 해석미분을 이용해서 문제를 해결할 수 없을 때 수치적 근사값을 이용해서 미분을 수행
# => 수치미분(Numenrcal derivative)
# delta_x라고 0과 근사한 값을 이용해야 한다
# delta_x를 소수점 8자리 이하로 내려가면 안된다(파이썬)
# 일반적으로 0.00001 수준으로 설정해서 수치미분 수행


# 수치미분방법 크게 3가지
# 전향차분, 후향차분, 중앙차분. 
# 중앙차분방식으로 미분을 수행(가장 정확한 값을 도출하기 때문)


# partial derivative(편미분)
# 입력변수가 하나 이상인 다변수 함수에서 미분하고자 하는 변수 하나를 제외한 
# 나머지 변수를 상수로 취급하고 해당 변수를 미분

# f(x,y) 에서 x에 대해 편미분 

# Chain rule(연쇄법칙)
# 여러 함수로 구성된 함수(합성함수 - composite function) 
# 합성함수를 미분하려면? => chain rule(연쇄법칙) 
# 합성함수를 구성하고 있는 각 함수를 미분한 후 그것들의 곱으로 표현

```

```python
## 일변수 함수의 수치미분 코드를 python으로 작성해보자

# 입력으로 들어오는 x에서 아주 미세하게 값이 변화할 때 함수 f는 
# 얼마나 변하는지 수치적으로 계산해서 return

# f(x) = x^2
# f'(5) = ?


def numerical_derivative(f, x):
    
    # f : 미분하려는 함수가 들어온다
    #     python은 일급함수를 지원한다
    #     하나의 함수를 다른 함수의 인자로 전달할 수 있다
    # 일반적으로 외부에서 def. lambda를 이용해서 정의.
    
    # x : 미분값을 알고자하는 입력값
    
    # delta_x는 0과 가까운 작은값을 이용하고 1e-8이하는 사용하면 안된다
    # 컴퓨터의 반올림오차때문에 엉뚱한 값이 나온다
    # delta_x는 1e-4 ~ 1e-6 정도의 수로 설정
    
    delta_x = 1e-4
    
    # 중앙차분으로 미분 수행
    return (f(x + delta_x) - f(x - delta_x)) / (2 * delta_x)

# 미분하려는 함수를 만들어요
# f(x) = x^2

def my_func(x):
    
    return x**2

# 함수 f(x) = x**2에서 미분계수 f'(5)를 구하세요

result = numerical_derivative(my_func,5)


print('result : {}'.format(result))
    
```

```python
# 다변수함수에 대한 미분
# 다변수 함수인 경우 입력변수가 하나 이상이므로, 
# 이 입력변수들은 서로 독립이기 때문에 수치미분할 때 개별적으로 계산해줘야 한다

# f(x,y) = 2x + 3xy = y^3
# f'(1.0,2.0) = ??

# 수치미분 최종코드

import numpy as np

def numerical_derivative(f,x):
    
    # f : 미분하려고 하는 다변수 함수
    # x : 모든 변수를 포함하고 있어야 한다. ndarray(차원 상관없이)
    
    delta_x = 1e-4
    derivative_x = np.zeros_like(x) # 미분한 결과를 저장하는 ndarray
    
    # ierator를 이용해서 입력변수 x에 대해 편미분을 수행
    
    it = np.nditer(x, flags=['multi_index'])
    
    while not it.finished:
        
        idx = it.multi_index  # iterator의 현재 index를 추출(tuple로 추출)
        
        # 현재 칸의 값을 어딘가에 잠시 저장
        tmp = x[idx]
        
        x[idx] = tmp + delta_x
        fx_plus_delta = f(x)  # f(x + delta_x)
        
        
        x[idx] = tmp - delta_x
        fx_minus_delta = f(x) # f(x - delta_x)
        
        derivative_x[idx] = (fx_plus_delta - fx_minus_delta) / (2 * delta_x)
        
        x[idx] = tmp # 데이터를 원상 복구
        
        it.iternext()
        
    
    return derivative_x


# 이렇게 구현된 최종 수치마분 함수를 이용해서
# 일변수 함수를 미분해 보자

# def my_func(x):
    
#     return x**2

## f(x) = x^2에서 f'(3) ??

# result = numerical_derivative(my_func,np.array([3.0]))
# print('결과 : {}'. format(result))
# 결과 : [6.]


# f(x,y) = 2x + 3xy +np.power(y,3)
# 이변수 함수를 미분

def my_func(input_data):
    x = input_data[0]
    y = input_data[1]
    
    return 2*x + 3*x*y + np.power(y,3)

result = numerical_derivative(my_func,np.array([1.0,2.0]))

# print('결과 : {}'. format(result))
# 결과 : [ 8.         15.00000001]
```

```python
# Linear Regression Model의 이론적인 부분을 python으로 구현해보자

# 코드 구현 절차

# 1. Training Data Set을 준비
#    머신러닝에 입력으로 사용될 데이터를 Numpy array(ndarray)형태로 준비

# 2. Linear Regression Model을 정의
#    y = Wx + b => model을 프로그램적으로 표현
#    w와 b에 대한 변수 선언한 후 초기값은 랜덤값을 이용할 것

# 3. loss function을 정의 : 손실함수(loss function)에 대한 코드를 작성
#                         matrix 처리해야 한다

# 4. learning rate의 정의 : 일반적으로 customixing이 되는 값으로 초기에는 
#                         0.001정도로 설정해서 사용하고 loss값을 보고 수치를 조절할 필요 있다

# 5. 학습을 진행 : 반복적으로 편미분을 이용해서 W와 b를 update하는 방식으로 구현


import numpy as np
import matplotlib.pyplot as plt

# Training Data Set
x_data = np.array([1,2,3,4,5,7,8,10,12,13,14,15,18,20,25,28,30]).reshape(-1,1)
t_data = np.array([5,7,20,31,41,44,46,49,60,62,70,80,85,91,92,97,98]).reshape(-1,1)


# 데이터의 분포를 scatter 확인

plt.scatter(x_data.ravel(), t_data.ravel())
plt.plot(x_data.ravel(), np.dot(x_data,W) + b)  # 직선
plt.show()


# Linear Regression Model을 정의

W = np.random.rand(1,1) # matrix

b = np.random.rand(1)   # scalar


# Hypothesis는 따로 표현하지 않는다

# loss function을 정의 


def loss_func(x,t):
    
    y = np.dot(x,W) + b  # Wx + b 
    
    return np.mean(np.power((t-y),2))  # 최소제곱법


# 미분함수

def numerical_derivative(f,x):
    
    # f : 미분하려고 하는 다변수 함수
    # x : 모든 변수를 포함하고 있어야 한다. ndarray(차원 상관없이)
    
    delta_x = 1e-4
    derivative_x = np.zeros_like(x) # 미분한 결과를 저장하는 ndarray
    
    # iterator를 이용해서 입력변수 x에 대해 편미분을 수행
    
    it = np.nditer(x, flags=['multi_index'])
    
    while not it.finished:
        
        # iterator의 현재 index를 추출(tuple로 추출)
        idx = it.multi_index  
        
        # 현재 칸의 값을 어딘가에 잠시 저장
        tmp = x[idx]
        
        x[idx] = tmp + delta_x
        fx_plus_delta = f(x)  # f(x + delta_x)
        
        
        x[idx] = tmp - delta_x
        fx_minus_delta = f(x) # f(x - delta_x)
        
        derivative_x[idx] = (fx_plus_delta - fx_minus_delta) / (2 * delta_x)
        
        x[idx] = tmp # 데이터를 원상 복구
        
        it.iternext()
        
    
    return derivative_x


# 학습 종료 후 임의의 데이터에 대한 예측값을 알아오는 함수
# prediction

def predict(x):
    
    return np.dot(x,W) + b   # Hypothesis, Linear Regression Model

# learning rate라는 상수가 필요. 정의해야 한다
learning_rate = 0.0001

# 미분을 진행할 loss_func()에 대한 lambda 함수를 정의

f = lambda x : loss_func(x_data, t_data)

# 학습을 진행
# 반복해서 학습을 진행 (W와 b를 update를 반복적으로 학습을 진행)

for step in range(90000):
    
    W = W - learning_rate * numerical_derivative(f,W)  # W의 편미분
    b = b - learning_rate * numerical_derivative(f,b)  # b의 편미분
    
    if step % 3000 == 0:
        print('W : {}, b : {}, loss : {}'.format(W,b,loss_func(x_data,t_data)))
        
        
# 학습종료 후 예측
print(predict(19))
```

```python
# Simple Linear Regression
# 온도에 따른 오존량 예측

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from my_library.machine_learning import numerical_derivative



################

def numerical_derivative(f,x):
    
    # f : 미분하려고 하는 다변수 함수
    # x : 모든 변수를 포함하고 있어야 한다. ndarray(차원 상관없이)
    
    delta_x = 1e-4
    derivative_x = np.zeros_like(x) # 미분한 결과를 저장하는 ndarray
    
    # iterator를 이용해서 입력변수 x에 대해 편미분을 수행
    
    it = np.nditer(x, flags=['multi_index'])
    
    while not it.finished:
        
        # iterator의 현재 index를 추출(tuple로 추출)
        idx = it.multi_index  
        
        # 현재 칸의 값을 어딘가에 잠시 저장
        tmp = x[idx]
        
        x[idx] = tmp + delta_x
        fx_plus_delta = f(x)  # f(x + delta_x)
        
        
        x[idx] = tmp - delta_x
        fx_minus_delta = f(x) # f(x - delta_x)
        
        derivative_x[idx] = (fx_plus_delta - fx_minus_delta) / (2 * delta_x)
        
        x[idx] = tmp # 데이터를 원상 복구
        
        it.iternext()
        
    
    return derivative_x


################

# 1. Row Data loading

df = pd.read_csv('./data/ozone.csv')
# display(df)


# 2. Data Preprecessing (데이터 전처리)
#    - 결측치 처리 
#     - 삭제, 값을 변경(평균, 최대, 최소) 값을 예측해서 값을 대체
#    - 이상치 처리(outlier)
#     - 이상치를 검출
#    - 데이터 정규화 작업
#     - 학습에 필요한 컬럼을 추출, 새로 생성


# 필요한 column(Temp, Ozone)만 추출
# 결치값을 제거

training_data = df[['Temp', 'Ozone']]
# print(training_data.shape)   # (153, 2)

training_data = training_data.dropna(how='any') # how = 'any' : 결측치가 있는 행을 다 지운다
print(training_data.shape)    # (116, 2)


# 3. Traning Data Set 준비
x_data = training_data['Temp'].values.reshape(-1,1)
t_data = training_data['Ozone'].values.reshape(-1,1)

# 4. Simple Linear Regression이기 때문에 
# y = Wx + b
# W, b 값 정의

W = np.random.rand(1,1) # x_data와 행렬곱해야하기 때문에 매트릭스 형태
b = np.random.rand(1)   # scalar

# 5. loss function 정의

def loss_func(x,t):
    
    y = np.dot(x, W) + b
    
    return np.mean(np.power((t-y),2)) # 최소제곱법


# 6. 학습종료 후 예측값을 알아오는 함수

def predict(x):
    
    return np.dot(x, W) + b

# 7. 기타 프로그램에서 필요한 변수들을 정의
learning_rate = 1e-6

f = lambda x : loss_func(x_data, t_data)


# 8. 학습 진행

for step in range(30000):
    
    W -= learning_rate * numerical_derivative(f, W)
    b -= learning_rate * numerical_derivative(f, b)
    
    if step % 3000 == 0:
        print('W : {}, b : {}, loss : {}'.format(W, b, loss_func(x_data, t_data)))
        

# 9. 그래프로 확인
plt.scatter(x_data, t_data)
plt.plot(x_data, np.dot(x_data,W)+ b, color = 'r')
plt.show()  


# 10. 예측
result = predict(62)
print(result)  # [[35.27133536]]
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from my_library.machine_learning import numerical_derivative





df = pd.read_csv('./data/ozone.csv')

training_data = df[['Temp','Ozone']]

training_data = training_data.dropna(how='any')

print(training_data.shape)


## 이상치 처리 ##
zscore_threshold = 2.0  # 97.7% 이상, 2.3% 이하인 값들을 이상치로 판별

# outlier 출력
# Temp에 대한 이상치(지대점)를 확인하고 제거
outliers = stats.zscore(training_data['Temp'])
outliers = training_data['Temp'][np.abs(stats.zscore(training_data['Temp'])) > zscore_threshold]

training_data = training_data.loc[~training_data['Temp'].isin(outliers)] # ~ : 뒤의 내용을 역으로 바꾸다

print(training_data.shape)



# outlier 출력
# Ozone에 대한 이상치(지대점)를 확인하고 제거
outliers = stats.zscore(training_data['Ozone'])
outliers = training_data['Ozone'][np.abs(stats.zscore(training_data['Ozone'])) > zscore_threshold]

training_data = training_data.loc[~training_data['Ozone'].isin(outliers)] # ~ : 뒤의 내용을 역으로 바꾸다/Oz

print(training_data.shape)


# Training Data Set

x_data = training_data['Temp'].values.reshape(-1,1)
t_data = training_data['Ozone'].values.reshape(-1,1)


# W, b

W = np.random.rand(1,1)
b = np.random.rand(1)


# loss function

def loss_func(x,t):
    
    y = np.dot(x,W) + b
    return np.mean(np.power((t-y),2))



def predict(x):
    
    return np.dot(x,W) + b


# learning_rate
learning_rate = 1e-5
f = lambda x : loss_func(x_data, t_data)

# 학습

for step in range(30000):
    
    W-= learning_rate * numerical_derivative(f,W)
    b-= learning_rate * numerical_derivative(f,b)
    
    if step % 3000 == 0 :
        print('W : {}, b : {}, loss : {}'.format(W,b,loss_func(x_data,t_data)))
        
        
plt.scatter(x_data, t_data)
plt.plot(x_data, np.dot(x_data,W)+ b, color = 'r')
plt.show()  


# 10. 예측
result = predict(62)
print(result)  # [[31.2191236]]
```

```python
import tensorflow as tf

# 1. Raw Data Loading
# 2. Data preprocessing(데이터전처리)


# 3. Training Data Set
x_data = [2,4,5,7,10]
t_data = [7,11,13,17,23]

# 4. W, b 정의
W = tf.Variable(tf.random.normal([1]), name='weight') # name: node에서 이름을 활용할 수 있음
b = tf.Variable(tf.random.normal([1]), name='bias')


# 5. Hypothesis, Simple Linear Regression Model
H = W * x_data + b

# 6. Loss function
loss = tf.reduce_mean(tf.square(t_data-H))

# 7. train node 생성
optimzer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
train = optimzer.minimize(loss)

# 8. 실행준비 및 초기화 작업
sess = tf.Session()
sess.run(tf.global_variables_initializer()) # 초기화 작업 (2.xx 에서는 사라졌음)

# 9. 반복해서 학습을 진행

for step in range(30000):
    _, W_val, b_val, loss_val = sess.run([train,W,b,loss])
    
    if step % 3000 == 0:
        print('W : {}, b : {}, loss : {}'.format(W_val,b_val,loss_val))
        
print(sess.run(H))
```
